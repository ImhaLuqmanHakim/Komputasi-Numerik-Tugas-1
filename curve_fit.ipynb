```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data sintetik dengan fungsi sinus
rng = np.random.RandomState(1)  # Seed untuk random state
x = 10 * rng.rand(50)  # Data x acak di rentang 0-10
y = np.sin(x) + 0.1 * rng.randn(50)  # Fungsi sinus dengan noise

# Plot data scatter
plt.scatter(x, y, label="Data", color="blue")
plt.title("Data Scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Definisi model fungsi (sinusoidal)
def model_func(x, a, b, c):
    return a * np.sin(b * x + c)

# Curve fitting
params, covariance = curve_fit(model_func, x, y)

# Menampilkan parameter hasil fitting
a_fit, b_fit, c_fit = params
print(f"Hasil Fitting: a = {a_fit:.3f}, b = {b_fit:.3f}, c = {c_fit:.3f}")

# Membuat prediksi menggunakan model hasil fitting
x_fit = np.linspace(0, 10, 500)  # Data x untuk prediksi
y_fit = model_func(x_fit, *params)  # Menghitung y berdasarkan hasil fitting

# Plot data dan hasil fitting
plt.scatter(x, y, label="Data", color="blue")
plt.plot(x_fit, y_fit, label="Model Fit", color="red", linewidth=2)
plt.title("Curve Fitting Result")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
